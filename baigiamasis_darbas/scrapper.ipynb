```python
from selenium import webdriver
import os 
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import requests
import time
from random import randint
from bs4 import BeautifulSoup
import pandas as pd
```

```python
options = Options()
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)

cars_data = []

for page_nr in range(1,201):
    url = f'https://m.autoplius.lt/skelbimai/naudoti-automobiliai?page_nr={page_nr}'
    driver.get(url)
    src = driver.page_source
    bs = BeautifulSoup(src, 'html.parser')
    skelbimai = bs.find_all('div', {'class':'description'})
    try:
        for skelbimas in skelbimai:
            metai = skelbimas.find('div', {'class':'title-year'}).text.strip()
            gamintojas = skelbimas.find('div', {'class':'title-content'}).text.strip()
            kaina = skelbimas.find('div', {'class':'pricing-container'}).find('strong').text.strip()
            parametrai = skelbimas.find('div', {'class':'item-parameters'}).text.strip().replace('\n\n','\n').replace('\n',';').replace('  ','')
            pars = parametrai.split(';')
            vieta = pars[-1]
            tipas = pars[-2]
            rida = 'Nenurodyta'
            if 'km' in parametrai:
                rida = pars[3]
            pavaros = 'Nenurodyta'
            if 'Mechaninė' or 'Automatinė' in parametrai:
                pavaros=pars[2]
            kuras = pars[0]
            variklis = 'Nenurodyta'
            if 'kW' in parametrai:
                variklis = pars[1]                
            print("====SKELBIMAS===")
            print(pars)
            d = {'gamintojas':gamintojas,'metai':metai,'kaina':kaina,'rida':rida,'pavaros':pavaros,'kuras':kuras,'variklis':variklis}
            cars_data.append(d)
    except Exception as ex:
        pass
    time.sleep(randint(1,5))
df = pd.DataFrame(data=cars_data)
df.to_csv('mobile_autoplius.csv', sep=';')     
```
