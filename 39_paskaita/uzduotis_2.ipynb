Jūsų užduoties tikslas pasiimti duomenis (nuskaityti failą, visus veiksmus atlikite vienam failui, tik po to kitam). Nustatyti, kuriuos stulpelius norite nagrinėti. Pasirinktu būdu sukurti klasterius. Po to atlikite tą patį su sekančiu failu ir palyginkite pasiskirstymą.

```python
data_2016 = pd.read_csv('Gyvensena2016.csv')

selected_features = data_2016[['fizine_veikla_be_pamoku', 'pusryciai_dd', 'pusryciai_savaitgaliai', 'vaisiai', 'darzoves', 'patirtos_patycios', 'fizines_bausmes', 'dantys', 'sveikata']]

agg_cluster = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

agg_cluster.fit(selected_features)

linked = linkage(selected_features, 'ward')
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchinis klasterizavimas')
plt.xlabel('Klasterių skaičius')
plt.ylabel('Atstumas')
plt.show()
```
