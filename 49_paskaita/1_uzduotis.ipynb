## 1 Užduotis
Naudokite šį duomenų rinkinį https://www.tensorflow.org/datasets/catalog/fashion_mnist
Transformuokite duomenis atsižvelgiant į tai, kokie duomenys yra pateikti duomenų rinkinyje (28px x 28px x 1 (spalvos kanalas))
Ištreniruokite savo modelį, patikrinkite jo tikslumą.

```python
import tensorflow as tf 
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Duomenų įkėlimas ir paruošimas
(train_images, train_labels), (test_images, test_labels) = datasets.fashion_mnist.load_data()
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

# Modelio apibrėžimas su teisinga įvesties forma
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Modelio kompiliavimas
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Modelio treniravimas
trained_model = model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Modelio vertinimas
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(test_acc)
```

```python
import tensorflow_datasets as tfds
```

```python
(ds_train, ds_test), ds_info = tfds.load(
    'cats_vs_dogs',
    split=['train', 'test'],
    with_info=True,
    as_supervised=True,
)
```

```python
from emnist import extract_training_samples, extract_test_samples

# Duomenų įkėlimas ir paruošimas
train_images, train_labels = extract_training_samples('balanced')
test_images, test_labels = extract_test_samples('balanced')

# Duomenų pertvarkymas ir normalizavimas
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1)).astype('float32') / 255

# Modelio apibrėžimas su teisinga įvesties forma
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(47, activation='softmax')
])

# Modelio kompiliavimas
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Modelio treniravimas
trained_model = model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Modelio vertinimas
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(test_acc)
```
